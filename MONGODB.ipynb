Rgex in filtering the data from db collection

```python
a = adh.find({"age":{"$gt":10}})
for x in a:
  print(x)
```

Greater - gt, Lesser  - lt

Or condition to filter data from db collection

```python
A = {"$or":[{"address":"coimbatore"},{"age":{"$gt":10}}]}
B = adh.find(A,{"_id":0})
for x in B:
  print(x)
```

Descending order

```python
mydoc = adh.find({"address":"coimbatore"}).sort("_id",-1)
for x in mydoc:
  print(x)
```

Regex coding with starting letter comparision

```python
A = adh.find({"address":{"$regex":"^c"}})
for x in A:
  print(x)
```

Regex coding with ending letter comparision with limit function. [limit function is used to limit the number of output]

```python
A = adh.find({"address":{"$regex":"de$"}}).limit(2)
for x in A:
  print(x)
```

Delete many from db collection

Update the existing file with a new key and data value

```python
A = {"name":"adharsh"}
new = {"$set":{"age":7}}
adh.update_one(A,new)
for x in adh.find():
  print(x)
```

Set value adds new key and value if it is not existing and if it is existing it will update it

Updating file using unset

```python
A = {"name":"adharsh"}
new = {"$unset":{"age":"7358"}}
adh.update_one(A,new)
for x in adh.find():
  print(x)
```

Updating time stamp

```python
A = {"name":"vaishag"}
new = {"$currentDate":{"lastmodified":True}}
adh.update_one(A,new)
for x in adh.find():
  print(x)
```
